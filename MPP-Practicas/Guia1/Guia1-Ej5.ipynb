```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.array([0,0.5,1.0,1.5,2,2.5,3])

z1 = np.linspace(0,3,num=7)

T = np.array([70,68,55,22,13,11,10])
```

```python
plt.plot(z,T)
plt.show()
```

```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
```

```python

cs = CubicSpline(z, T)

xs = np.arange(0,3.1, 0.1)

fig, ax = plt.subplots(figsize=(6.5, 4))

ax.plot(z, T, 'o', label='data')


ax.plot(xs, cs(xs), label="S")

# Quiero ver las sucesivas derivadas que tiene la funcion

ax.plot(xs, cs(xs, 1), label="S'")

ax.plot(xs, cs(xs, 2), label="S''")

#ax.plot(xs, cs(xs, 3), label="S'''")

ax.legend(loc='lower left', ncol=2)
plt.show()
```

```python
t2 = cs(xs, 1)

idx_0 = np.argmin(t2)

print(xs[idx_0])
```

```python
k = 0.01
j = - k *t2

fig, ax = plt.subplots()
ax.plot(xs,j)
plt.show()
```

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
# Datos del problema 

m = 70      # [kg]
r = 12.5    # [kg/s]
g = 10      # [m/s2]

#......................................................................................

#           PARA DEFINIR LAS FUNCIONES QUE VOY A NECESITAR

def solDiscreta (l_inf,l_sup,intervalo):

    t = np.arange(l_inf,l_sup,intervalo)
    v_dis = np.zeros(int((l_sup-l_inf)/intervalo))
    
    for i in range(0,v_dis.size-1,1):

        v_dis[i+1] = v_dis[i] + (g - r/m * v_dis[i]) * intervalo

    return t,v_dis


def solDiscreta2 (l_inf,l_sup,intervalo):

    t = np.arange(l_inf,l_sup,intervalo)
    v_dis = np.zeros(int((l_sup-l_inf)/intervalo))
    
    for i in range(0,v_dis.size-1,1):

        v_dis[i+1] = v_dis[i] + (g - 0.25/m * (v_dis[i])**2) * intervalo

    return t,v_dis

#...................................................................................

# Se debe evaluiar la velocidad del paracaidista 

# Defino mi vector t

limT = 30

t = np.arange(0,limT,2)    # Este el vector de tiempo ...  Se calculan intervalos de 2 seg


# Esta es la expresion exacta para la resolucion del problema - Teoria_1

v = (g*m)/r * ( 1 - np.exp( -(r/m) * t ))

vlim = g*m/r

# Esta es la solucion discreta para la resolucion del problema - Teoria 1

t1,v_dis1 = solDiscreta (0,limT,1)
#t2,v_dis2 = solDiscreta (0,limT,1)
#t3,v_dis3 = solDiscreta (0,limT,0.5)

t4,v_dis4 = solDiscreta2 (0,limT,1)

#.....................................................................................

#           PARA HACER LOS GRAFICOS 

plt.figure(figsize=(12, 6))
  
plt.plot(t,v,label="Solucion exacta", color="b", marker="none")
plt.plot(t1,v_dis1,label="Solucion discreta (2 seg)", color="g",linestyle="--", marker="o")
#plt.plot(t2,v_dis2,label="Solucion discreta (1 seg)", color="r",linestyle="--", marker="o")
#plt.plot(t3,v_dis3,label="Solucion discreta (0.5 seg)", color="y",linestyle="--", marker="o")
plt.axhline(y=vlim, color="r", linestyle="--", label=f"Vel. lmite = {vlim:.2f} m/s")

plt.plot(t4,v_dis4,label="Solucion discreta (2 seg)", color="c",linestyle="--", marker="o")

plt.title("Velocidad del Paracaidista en funcin del tiempo")
plt.xlabel("Tiempo [s]")
plt.ylabel("Velocidad [m/s]")
plt.legend()
plt.grid(True)

plt.show()

```

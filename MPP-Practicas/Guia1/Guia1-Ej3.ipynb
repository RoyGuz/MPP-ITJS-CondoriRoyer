```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
# Datos del problema 

m = 70      # [kg]
r = 12.5    # [kg/s]
g = 10      # [m/s2]

# Se debe evaluiar la velocidad del paracaidista 

# Defino mi vector t

t = np.arange(0,50,2)    # Este el vector de tiempo ... 

v = (g*m)/r * ( 1 - np.exp( -(r/m) * t ))
               
plt.plot(t,v)


```

```python
import numpy as np
```

```python
A = np.array([
    [1, -3, -2],
    [2, -4, -3],
    [-3, 6, 8]
])

B = np.array([6, 8, -5])

X = np.linalg.solve(A,B)    # Comando para resolver sistemas lineales   

print(A)
```

```python
np.allclose(np.dot(A,X),B)  # Comando para verificar si el sistema esta correctamente resuelto
```

Esta sera la primer prueba para un problema unidimencional

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
#   La ecuacion diferencial de una barra de longitud L que transmite calor
#   entre dos fuentes de temperatura conocida T1 y T2 es:

#   --- d^2T/dx^2 = h'(T - Ta)---

#   Siendo:
#   Ta: Temperatura ambiente

#   --- h' = ( c * p )/( k * A ) = hp --- 
#   Donde: 
#   c: Coeficiente de conveccion entre la barra y el ambiente
#   p: Perimetro de la barra
#   k: Coeficiente de transmicion de calor
#   A: Seccin de la barra

#   Resolver la ecuacion propuesta usando el metodo de diferencias finitas
#   en el intervalo [0 L]

#...............................................................................................

#   Defino mis datos:

TI = 50     # Temperatura en un extremo x = 0
TD = 10     # Temperatura en un extremo x = L

Ta = 20     # Temperatura ambiente 

hp = 0.2    # Coeficiente h'

L = 10      # Longitud de la barra

#................................................................................................

# Lo primero que se hce es discretizar el problema 

N = 20     # Defino la cantida de intervalos

paso = L/N  # Se define el paso

aux1 = 2 + hp*paso**2       # Este es el valor de la diagonal principal
aux2 = hp * paso**2 * Ta    # Este es el valor que va a estar en el vector B
```

```python
# Primero armo una matriz

A = np.zeros((N+1, N+1))

#   Armo y completo la matriz

np.fill_diagonal(A, aux1)
np.fill_diagonal(A[1:], -1)
np.fill_diagonal(A[:, 1:], -1)

#   Corrijo los valores 

A[0,0] = 1
A[-1,-1] = 1

A[-1,-2] = 0
A[0,1] = 0

#................................................................................

# Armo el vectr B

B = np.zeros((N+1,1))
B[:] = aux2
B[0] = TI
B[-1] = TD
```

```python
X = np.linalg.solve(A,B)    # Comando para resolver sistemas lineales   

print(X)
```
